```python
import nltk
from transformers import pipeline

nltk.download('punkt')

# Zero-shot classification pipeline
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0  # use GPU if available
)

# Candidate labels for the type of message, refined to capture nuance
candidate_labels = ["Challenging the premise", "Neutral/Informative", "Supportive"]
```

```python
def classify_message(response):
    result = classifier(response, candidate_labels=candidate_labels, multi_label=False)
    # Return top label
    return result['labels'][0], result['scores'][0]  # label and confidence score
```

```python

# Apply classification to each response with the new labels
df[['message_label', 'confidence']] = df['response'].apply(lambda x: pd.Series(classify_message(x)))

pd.set_option('display.max_colwidth', None)
display(df)
```
