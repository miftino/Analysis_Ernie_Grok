```python
import nltk
from transformers import pipeline

nltk.download('punkt')

# Zero-shot classification pipeline
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0  # use GPU if available
)

# Candidate labels for the type of message, refined to capture nuance
candidate_labels = ["Challenging the premise", "Neutral/Informative", "Supportive"]
```

```python
def classify_message(response):
    result = classifier(response, candidate_labels=candidate_labels, multi_label=False)
    # Return top label
    return result['labels'][0], result['scores'][0]  # label and confidence score
```

```python
# Apply classification to each response with the new labels
df[['message_label', 'confidence']] = df['response'].apply(lambda x: pd.Series(classify_message(x)))

pd.set_option('display.max_colwidth', None)
display(df)
```

```python
comparison_df = df.groupby(['bot', 'message_label']).size().unstack(fill_value=0)
display(comparison_df)
```

## Project Notes: Bot Response Analysis

This notebook is set up to analyze and compare responses from different AI bots (Grok and ERNIE) to geopolitical questions, focusing on how they frame their answers.

### Current Analysis Method:

We are using a **Zero-Shot Classification** model (`facebook/bart-large-mnli`) to categorize bot responses without requiring pre-labeled training data. The model determines how likely a response is to 'entail' a given label.

### Current Classification Labels:

To capture the nuanced way bots might approach a question, we are using the following `candidate_labels`:

*   **Challenging the premise:** Indicates if the bot's response questions the underlying assumption, implication, or simplicity of the user's question, rather than giving a direct answer.
*   **Neutral/Informative:** For responses that provide factual, objective information without taking a strong stance or questioning the premise.
*   **Supportive:** For responses that acknowledge or affirm the question's premise before providing an answer.

### Interpreting the Results (`comparison_df`):

The `comparison_df` table summarizes the counts of each `message_label` for Grok and ERNIE. This table is key for your research:

*   **Quantitative Comparison:** By observing the counts, you can identify trends (e.g., if one bot consistently challenges premises more often) as you expand your dataset with more questions.
*   **Bias Detection:** Significant differences in these categorizations between bots can highlight potential differences in their underlying programming, training data, or perceived national perspectives.
*   **Qualitative Deep Dive:** When a trend emerges, you can revisit specific responses to understand *how* a bot challenges a premise or offers support, providing rich qualitative data for your research.
